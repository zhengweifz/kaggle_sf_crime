```python
import os
import seaborn as sns
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
```

```python
train_df = pd.read_csv("train.csv")
test_df = pd.read_csv("test.csv")
```

```python
test_df["Category"] = None
#these two column are not in test set
train_df.drop(["Descript", "Resolution"], axis=1, inplace=True)
```

```python
train_df["IsTrain"] = 1
test_df["IsTrain"] = 0
whole_df = pd.concat([train_df, test_df])
```

```python
#time dimension
dates = pd.to_datetime(whole_df["Dates"])
whole_df["Year"] = dates.dt.year
whole_df["Month"] = dates.dt.month
whole_df["Day"] = dates.dt.day
whole_df["Hour"] = dates.dt.hour
whole_df["Minute"] = dates.dt.minute
whole_df["Weekday"] = dates.dt.weekday
whole_df["Quarter"] = dates.dt.quarter
whole_df["DayLight"]=0
whole_df["DayLight"][(whole_df["Month"] == 1) & (whole_df["Hour"] > 9) & (whole_df["Hour"] < 18)] = 1
whole_df["DayLight"][(whole_df["Month"] == 2) & (whole_df["Hour"] > 8) & (whole_df["Hour"] < 19)] = 1
whole_df["DayLight"][(whole_df["Month"] == 3) & (whole_df["Hour"] > 7) & (whole_df["Hour"] < 19)] = 1
whole_df["DayLight"][(whole_df["Month"] == 4) & (whole_df["Hour"] > 6) & (whole_df["Hour"] < 20)] = 1
whole_df["DayLight"][(whole_df["Month"] == 5) & (whole_df["Hour"] > 6) & (whole_df["Hour"] < 20)] = 1
whole_df["DayLight"][(whole_df["Month"] == 6) & (whole_df["Hour"] > 6) & (whole_df["Hour"] < 21)] = 1
whole_df["DayLight"][(whole_df["Month"] == 7) & (whole_df["Hour"] > 6) & (whole_df["Hour"] < 21)] = 1
whole_df["DayLight"][(whole_df["Month"] == 8) & (whole_df["Hour"] > 6) & (whole_df["Hour"] < 21)] = 1
whole_df["DayLight"][(whole_df["Month"] == 9) & (whole_df["Hour"] > 7) & (whole_df["Hour"] < 20)] = 1
whole_df["DayLight"][(whole_df["Month"] == 10) & (whole_df["Hour"] > 7) & (whole_df["Hour"] < 20)] = 1
whole_df["DayLight"][(whole_df["Month"] == 11) & (whole_df["Hour"] > 7) & (whole_df["Hour"] < 19)] = 1
whole_df["DayLight"][(whole_df["Month"] == 12) & (whole_df["Hour"] > 8) & (whole_df["Hour"] < 18)] = 1
whole_df["HourInterval"]=pd.cut(whole_df["Hour"] ,[0, 3, 6, 9, 12, 15, 18, 21, 24],include_lowest=True)
whole_df["Weekend"] = 0
whole_df["Weekend"][whole_df["Weekend"].isin([0,6])] = 1

```

```python
#location
whole_df["Intersection"] = 0
whole_df["Intersection"][whole_df["Address"].str.contains("/")] = 1
pca = PCA(n_components=1)
xy = whole_df[["X", "Y"]].values
pca.fit(xy)
whole_df["XY_PCA"] = pca.transform(xy)
```
